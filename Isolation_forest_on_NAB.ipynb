```python
from sklearn.ensemble import IsolationForest
#to_model_column='value'
clf=IsolationForest(n_estimators=10, max_samples='auto', 
contamination=float(.04), \
                        max_features=1.0, bootstrap=False, n_jobs=-1, random_state=42, verbose=0)
clf.fit(full_df[['value']])
```

```python
df1['scores']=clf.decision_function(df1[['value']])
df1['anomaly']=clf.predict(df1[['value']])
df1.head(100)
```

```python
df1
```

```python
df1.loc[df1['anomaly'] == 1,'anomaly'] = 0
df1.loc[df1['anomaly'] == -1,'anomaly'] = 1
df1.value_counts()
```

```python
test = df1['IsAnomaly']
pred = df1['anomaly']
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(test,pred))
```

```python
print(cl(classification_report(test, pred), attrs = ['bold']))
```

```python
from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(test,pred))
```
